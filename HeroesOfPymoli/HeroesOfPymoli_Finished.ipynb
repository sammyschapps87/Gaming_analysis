* Establish bins for ages

```python
#AGE
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
bin_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age_Bins"] = pd.cut(purchase_data["Age"],bins=bins,labels=bin_labels)
purchase_data.head()

purchasing_analysis_age_group = purchase_data.groupby("Age_Bins")

purchasing_analysis_age_group_count = purchasing_analysis_age_group['SN'].nunique()
purch_analy_age = pd.DataFrame(purchasing_analysis_age_group_count)
purch_analy_age['Percentage of Players'] = purch_analy_age['SN']/purch_analy_age['SN'].sum()
purch_analy_age['Percentage of Players'] = purch_analy_age['Percentage of Players'].map("{:.2%}".format)

purch_analy_age
```

```python
# PA AGE

purch_analy_age['Purchase Count'] = purchasing_analysis_age_group['Item ID'].count()
purch_analy_age['Average Purchase Price'] = purchasing_analysis_age_group['Price'].mean()
purch_analy_age['Total Purchase Value'] = purch_analy_age['Purchase Count']*purch_analy_age['Average Purchase Price']
# purch_analy_age_df['AvgTotalPurchasePrice'] = purchasing_analysis_age_group['Price']/purch_analy_age_df['Purchase Count']
purch_analy_age['Average Total Purchase per Person'] = purchasing_analysis_age_group['Price'].sum()/purchasing_analysis_age_group['SN'].nunique()

purch_analy_age['Average Purchase Price'] = purch_analy_age['Average Purchase Price'].map("${:,.2f}".format)
purch_analy_age['Total Purchase Value'] = purch_analy_age['Total Purchase Value'].map("${:,.2f}".format)
purch_analy_age['Average Total Purchase per Person'] = purch_analy_age['Average Total Purchase per Person'].map("${:,.2f}".format)

purch_analy_age[['Purchase Count','Average Purchase Price','Total Purchase Value','Average Total Purchase per Person']]
```

```python
#TOP SPENDER 
top_spenders_pre = purchase_data.groupby(['SN'])

purchase_count_tp = top_spenders_pre['Price'].count()
average_purchase_price_tp = top_spenders_pre['Price'].mean()
total_purchase_value_tp = average_purchase_price_tp*purchase_count_tp

top_spenders_df = pd.DataFrame({"Purchase Count": purchase_count_tp, "Average Purchase Price": average_purchase_price_tp, "Total Purchase Value": total_purchase_value_tp})

top_spenders_df = top_spenders_df.sort_values('Total Purchase Value', ascending=False)

top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map("${:,.2f}".format)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map("${:,.2f}".format)

top_spenders_df.head()
```

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting

```python
# MOST POP

most_pop_pre = purchase_data[['Item ID', 'Item Name', 'Price']]
most_pop_pre2 = most_pop_pre.groupby(['Item ID','Item Name'])

purchase_count_mp = most_pop_pre2['Price'].count()
item_price_mp = most_pop_pre2['Price'].mean()
total_purchase_value_mp = item_price_mp * purchase_count_mp

most_popular_df = pd.DataFrame({"Purchase Count": purchase_count_mp, "Item Price": item_price_mp, "Total Purchase Value": total_purchase_value_mp})

most_popular_df['Item Price'] = most_popular_df['Item Price'].map("${:,.2f}".format)
most_popular_df['Total Purchase Value'] = most_popular_df['Total Purchase Value'].map("${:,.2f}".format)

most_popular_df = most_popular_df.sort_values('Purchase Count', ascending=False)

most_popular_df.head()
```

## Most Profitable Items

```python
most_prof_pre = purchase_data[['Item ID', 'Item Name', 'Price']]
most_prof_pre2 = most_prof_pre.groupby(['Item ID','Item Name'])

purchase_count_mp = most_prof_pre2['Price'].count()
item_price_mp = most_prof_pre2['Price'].mean()
total_purchase_value_mp = item_price_mp * purchase_count_mp

most_prof_df = pd.DataFrame({"Purchase Count":purchase_count_mp,"Item Price":item_price_mp,"Total Purchase Value":total_purchase_value_mp})

most_prof_df = most_prof_df.sort_values('Total Purchase Value', ascending=False)

most_prof_df['Item Price'] = most_prof_df['Item Price'].map("${:,.2f}".format)
most_prof_df['Total Purchase Value'] = most_prof_df['Total Purchase Value'].map("${:,.2f}".format)



most_prof_df.head()
```
